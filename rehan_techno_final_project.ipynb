# Detecting Outlier 

```python
# Define the columns to be plotted

columns = ['CreditScore', 'MIP', 'Units', 'OCLTV', 'DTI', 'OrigUPB', 
           'LTV', 'OrigInterestRate', 'OrigLoanTerm', 'MonthsDelinquent', 'MonthsInRepayment']
```

```python
# Create boxplots for each numerical column before handling outliers

plt.figure(figsize=(16, 10))
for i, column in enumerate(columns, start=1):
    plt.subplot(3, 4, i)
    sns.boxplot(df[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

# Handle Outlier and Capping them

```python
# Handle outliers using IQR and capping them

for column in columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap outliers
    df[column] = df[column].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```

```python
print("Shape of DataFrame after handling outliers:", df.shape)
```

# After handling Outlier

```python
# Create boxplots for each numerical column after handling outliers

plt.figure(figsize=(16, 10))
for i, column in enumerate(columns, start=1):
    plt.subplot(3, 4, i)
    sns.boxplot(df[column])
    plt.title(f"After: {column}")

plt.tight_layout()
plt.show()
```
